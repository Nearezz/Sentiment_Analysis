```python
import pandas as pd

# Load cleaned datasets
tweets = pd.read_pickle("tweets_high_impact.pkl")
prices = pd.read_pickle("price_clean.pkl")

# Ensure proper datetime
tweets["date"] = pd.to_datetime(tweets["date"]).dt.date
prices["date"] = pd.to_datetime(prices["date"]).dt.date

# MERGE (many tweets â†’ one price row)
merged = pd.merge(
    tweets,
    prices,
    on=["ticker", "date"],
    how="left"
)

merged = merged.drop(columns=[
    "favorite_count",
    "is_retweet",
    "retweet_count",
    "retweets",
    "likes",
    "is_noise"
], errors="ignore")

merged.shape, merged.head()
```

```python
merged.to_pickle("merged_dataset.pkl")
merged.to_csv("merged_dataset.csv", index=False)
```
